## N-Atomic Function for 1D

### Derivations

#### Monoatomic

The forces connecting any 2 atoms in a chain can be represented by a classical spring force $F = -C_NX_s$ where the displacement, $X_s = sa$ with $s$ being the unit cell position and $a$ being the spacing between unit cells.  
For each atom in the chain there are phonon vibrations that follow the general wave equation $A_s = Ae ^{i(ksa-{\omega}t)}$.  
When you move into new unit cells the variable $s$ changes as a function $s{\pm}n$ where $n$ is the displacement from the origin unit cell.  
For a monoatomic chain any connection is to a new unit cell which has a plane wave of the form 
\begin{equation} 
A_s \pm 1 = Ae^{i(k(s \pm 1)a-\omega t)}  
= A_s e^{\pm ika} \tag{1}
\end{equation} 

In the case of the monoatomic chain both nearest neighbors are new unit cells. The resulting net forces on the observed atom are represented by
\begin{equation}
F_s = -C(U_s - U_{s-1}) -C (U_s - U_{s+1}) \tag{2}
\end{equation}

With the formalism of {Eq1} plugged in then {Eq2} simplifies into
\begin{equation}
F_s = -C(U_s - U_s e^{-ika}) -C (U_s - U_s e^{ika}) 
= C(e^{ika} + e^{-ika}-2)U_s \tag{3}
\end{equation}

 In this case the term $e^{ika} + e^{-ika}$ can be simplified first by using Euler's Identity for
 \begin{equation}
 e^{ika} + e^{-ika} = 2cos(ka)
 \end{equation}

 Giving the force expression
 \begin{equation}
F_s = 2C(cos(ka)-1)U_s \tag{4}
\end{equation}

Finally before solving for $\omega ^2$ the term $1-cos(ka)$ can be further simplified with a double angle identity as
 \begin{equation}
 1-cos(ka) = 2sin^2 (ka/2)
 \end{equation}

 Giving the final general form
  \begin{equation}
F_s = -4Csin^2 (ka/2)U_s \tag{5}
\end{equation}

But this just describes the general form of the net force on the atom. To actually observe the system in a realistic sense the force needs to be compared to some part of the phononic wave that can be observed easily.
To further analyze this system a step must be taken slightly back to analyze what $F_s $ represents for each atom.  
Using Newton's 2nd Law it is well known that $F_s = m a$ where the acceleration is the second derivative of the position equation, which in this case is the plane wave from $Eq1$.
So now there is the form 
\begin{equation}
F_s = m a 
 = m \frac{\delta ^2 A_s}{\delta t^2}  
 = -m \omega ^2 A_s \tag{6}
\end{equation}

By combining {Eq5} and {Eq6} the net force is now described as
\begin{equation}
-m \omega ^2 U_s = -4Csin^2 (ka/2)U_s \tag{7}
\end{equation}

Which can be solved in terms of $\omega ^2$ to determine the dispersion relation of the phonons for each wave vector, $k$.
\begin{equation}
 \omega ^2  = \frac{4C}{m} sin^2 (ka/2) \tag{8}
\end{equation}

And which can finally be solved in terms of $\omega$ as a function of $k$ to plot the phonon dispersion
\begin{equation}
 \omega = 2\sqrt{\frac{C}{m}} sin(ka/2) \tag{9}
\end{equation}

#### Diatomic

The forces connecting any 2 atoms in a chain can be represented by a classical spring force $F = -C_NX_s$ where the displacement, $X_s = sa$ with $s$ being the unit cell position and $a$ being the spacing between unit cells.  
For each atom in the chain there are phonon vibrations that follow the general wave equation $A_s = Ae ^{i(ksa-{\omega}t)}$.  
When you move into new unit cells the variable $s$ changes as a function $s{\pm}n$ where $n$ is the displacement from the origin unit cell.  
For a diatomic unit cell any connection to a new pair leads to a new unit cell which has a plane wave of the form 

\begin{equation} 
A_s \pm 1 = Ae^{i(k(s \pm 1)a-\omega t)}  
= A_s e^{\pm ika} \tag{1}
\end{equation} 

To determine the actual motion of the phonons the net forces on the system need to be evaluated for each mass in the system.

\begin{equation}
F_s ^{m_1} = -C_2 (U_s - V_{s-1}) -C_1 (U_s - V_s) \tag{2}
\end{equation}
\begin{equation}
F_s ^{m_2} = -C_1 (V_s - U_s) -C_2 (V_s - U_{s+1}) \tag{3}
\end{equation}

Now when you take the general form for new supercells from $Eq1$ and apply it to $Eq2$ and $Eq3$ the new form becomes...
\begin{equation}
F_s ^{m_1} = -C_2 (U_s - V_s e^{-ika}) -C_1 (U_s - V_s) \tag{4}
\end{equation}
\begin{equation}
F_s ^{m_2} = -C_1 (V_s - U_s) -C_2 (V_s - U_s e^{ika}) \tag{5}
\end{equation}

To further analyze this system a step must be taken slightly back to analyze what $F_s ^{m_n}$ represents for each atom.  
Using Newton's 2nd Law it is well known that $F_s ^{m_n} = m_n a$ where the acceleration is the second derivative of the position equation, which in this case is the plane wave from $Eq1$.
So now there is the form 
\begin{equation}
F_s ^{m_n} = m_n a 
 = m_n \frac{\delta ^2 A_s}{\delta t^2}  
 = -m_n \omega ^2 A_s \tag{6}
\end{equation}

With the form of $Eq6$ subbed into $Eq4$ and $Eq5$ and sorted based on atoms an order form can be created as

\begin{equation}
0 = [m_1 \omega ^2 -(C_1 + C_2)]U_s + (C_1 + C_2 e^{-ika} )V_s \tag{7}
\end{equation}
\begin{equation}
0 = (C_1 + C_2 e^{ika})U_s + [m_2 \omega ^2 -(C_1 + C_2)]V_s \tag{8}
\end{equation}

To properly view these phonons the wave velocity $\omega$ needs to be understood in terms of the wave vectors, $k$ allowed within the system. To find solutions for $\omega$ a matrix is formed and the determinant can be found as 

$$
det
\begin{bmatrix}
m_1 \omega ^2 -(C_1 + C_2) & C_1 + C_2 e^{-ika} \\
C_1 + C_2 e^{ika} & m_2 \omega ^2 -(C_1 + C_2)
\end{bmatrix}
\begin{bmatrix}
U_s \\
V_s
\end{bmatrix}
=0 \tag{9}
$$

 to give solutions for $\omega ^2$.  
 Now the determinant for this matrix evaluates to 
 \begin{equation}
 m_1 m_2 \omega ^4 - (C_1 + C_2)(m_1 + m_2) \omega ^2 + 2C_1 C_2 - C_1 C_2 (e^{ika} + e^{-ika}) = 0 \tag{10}
 \end{equation}

 In this case the term $e^{ika} + e^{-ika}$ can be simplified first by using Euler's Identity for
 \begin{equation}
 e^{ika} + e^{-ika} = 2cos(ka)
 \end{equation}

 which bring the overall equation to the form
\begin{equation}
 m_1 m_2 \omega ^4 - (C_1 + C_2)(m_1 + m_2) \omega ^2 + 2C_1 C_2 (1 - cos(ka)) = 0 \tag{11}
\end{equation}

Finally before solving for $\omega ^2$ the term $1-cos(ka)$ can be further simplified with a double angle identity as
 \begin{equation}
 1-cos(ka) = 2sin^2 (ka/2)
 \end{equation}

 such that the final form of the determinant is 
\begin{equation}
 m_1 m_2 \omega ^4 - (C_1 + C_2)(m_1 + m_2) \omega ^2 + 4C_1 C_2 sin^2 (ka/2) = 0 \tag{12}
\end{equation}

To solve for $\omega ^2$ this is simply a quadratic equation where we can plug in values such that
\begin{equation}
\omega ^2 = \frac{(C_1 + C_2)(m_1 + m_2) \pm \sqrt{[(C_1 + C_2)(m_1 + m_2)]^2 - 16m_1 m_2 C_1 C_2 sin^2 (ka/2)}}{2 m_1 m_2} \tag{13}
\end{equation}

By treating solving the differing parts $\pm$ as the Optical and Acoustic  branches and solving for purely $\omega$ over a range of wave vectors, $k$ the solutions can be found and plotted.

### Triatomic and Tetratomic

Any N atomic system greater than $N=2$ will follow a process similar to the diatomic system, but different in that there are now atoms isolated entirely within the home unit cell $s$ which does not branch into any cells $s \pm n$. This leads to some "emptiness" within the determinant matrix as some atoms simply do not directly interact. This is first seen with Tetratomic systems which will be covered after a brief look at a Triatomic array.  
The force equations for a triatomic system take the forms
\begin{equation}
F_s ^{m_1} = -C_3 (U_s - Q_{s-1}) -C_1 (U_s - V_s) \tag{1}
\end{equation}
\begin{equation}
F_s ^{m_2} = -C_1 (V_s - U_s) -C_2 (V_s - Q_s) \tag{2}
\end{equation}
\begin{equation}
F_s ^{m_3} = -C_2 (Q_s - V_s) -C_3 (Q_s - U_{s+1}) \tag{3}
\end{equation}

Taking into account the wave form of the net force using Newton's 2nd Law and sorting the equations based on wave functions leads to
\begin{equation}
0 = [m_1 \omega ^2 - (C_3 + C_1)] U_s + C_1 V_s +C_3 e^{-ika} Q_s \tag{4}
\end{equation}
\begin{equation}
0 = C_1 U_s - [m_2 \omega ^2 - (C_1 +C_2)] V_s + C_2 Q_s \tag{5}
\end{equation}
\begin{equation}
0 = C_3 e^{ika} U_s + C_2 V_s + [m_3 \omega ^2 - (C_2 + C_3)] Q_s \tag{6}
\end{equation}

which can then be placed into a matrix equation of the form
$$
det
\begin{bmatrix}
m_1 \omega ^2 - (C_3 + C_1) & C_1 &  C_3 e^{-ika} \\
C_1 & m_2 \omega ^2 - (C_1 +C_2) & C_2 \\
C_3 e^{ika} & C_2 & m_3 \omega ^2 - (C_2 + C_3)
\end{bmatrix}
\begin{bmatrix}
U_s \\
V_s \\
Q_s
\end{bmatrix}
=0 \tag{7}
$$

The final simplification of the determinant comes down to
\begin{equation} 
m_1 m_2 m_3 \omega ^6 -[m_1 m_2 (C_2 +C_3) + m_1 m_3 (C_1 +C_3) + m_2 m_3 (C_1 +C_3)] \omega ^4 +[(C_1 C_2 + C_1 C_3 + C_2 C_3)(m_1 + m_2 + m_3)] \omega ^2
\tag{8}
\end{equation}
\begin{equation}
 - 4 C_1 C_2 C_3 sin^2 (ka/2)=0   
\tag{8}
\end{equation}

Which when solving for $\omega ^2$ is a cubic function that cannot easily be solved analytically. That is where the calculations of this package come into play.  


Now what about the case for a Tetratomic system with 4 different masses?  
The initial force equations follow the same pattern as the Triatomic system with there being fully "internal" atoms stuck within the unit cell.
\begin{equation}
F_s ^{m_1} = -C_4 (U_s - N_{s-1}) -C_1 (U_s - V_s) \tag{8}
\end{equation}
\begin{equation}
F_s ^{m_2} = -C_1 (V_s - U_s) -C_2 (V_s - M_s) \tag{9}
\end{equation}
\begin{equation}
F_s ^{m_3} = -C_2 (M_s - V_s) -C_3 (M_s - N_s) \tag{10}
\end{equation}
\begin{equation}
F_s ^{m_4} = -C_3 (N_s - M_s) -C_4 (N_s - U_{s+1}) \tag{11}
\end{equation}

Sorting by the wave functions gives the form
\begin{equation}
0 = [m_1 \omega ^2 - (C_4 + C_1)] U_s + C_1 V_s +0 M_s +C_4 e^{-ika} Ns \tag{12}
\end{equation}
\begin{equation}
0 = C_1 U_s + [m_2 \omega ^2 - (C_1 +C_2)] V_s + C_2 M_s + 0 N_s \tag{13}
\end{equation}
\begin{equation}
0 = 0 U_s + C_2 V_s + [m_3 \omega ^2 - (C_2 + C_3)] M_s + C_3 N_s \tag{14}
\end{equation}
\begin{equation}
0 = C_4 e^{ika} U_s + 0 V_s + C_3 M_s + [m_3 \omega ^2 - (C_3 + C_4)] N_s \tag{15}
\end{equation}

which can then be placed into a matrix equation of the form
$$
det
\begin{bmatrix}
m_1 \omega ^2 - (C_4 + C_1) & C_1 & 0 & C_4 e^{-ika} \\
C_1 & m_2 \omega ^2 - (C_1 +C_2) & C_2 & 0 \\
0 & C_2 & m_3 \omega ^2 - (C_2 + C_3) & C_3 \\
C_4 e^{ika} & 0 & C_3 & m_3 \omega ^2 - (C_3 + C_4)
\end{bmatrix}
\begin{bmatrix}
U_s \\
V_s \\
M_s \\
N_s
\end{bmatrix}
=0 \tag{16}
$$

Which is again a system with a tedious derivation of the determinant which cannot be easily solved analytically. However, this in tandem with the Triatomic system shows a pattern that begins to occur within the multi-atomic systems. Namely the values of the diagonal, 2nd diagonal, and off diagonals follow a strict pattern that expands based on the number of $N$ atoms present in the chain.  
Whether this holds for 2D systems is still to be seen.

### Actual Calculation of Eigenvalues

In general for a matrix system in which you want to find the eigenstates (ie the zero points of a system that remain constant under change) you will solve a problem of the form
\begin{equation}
A \nu = \lambda \nu
\end{equation}
where to solve the eigenvalues the system would be arranged as
\begin{equation}
det|A - \lambda I|\nu = 0
\end{equation}

however that is not quite the form being observed. Rather the problems being observed follow a more general form of
\begin{equation}
F \nu = \lambda M \nu
\end{equation}

Where $F=A$ as the general force equation for variables that fall under the vectors in $\nu$. Futhermore, there are not just unknown eigenvalues, $\lambda$, to solve for, but also the mass matrix, $M$, which is present.  
Now in the case of an written solution one could think to solve this by doing 
\begin{equation}
M^{-1} F \nu = \lambda \nu
\end{equation}
and creating the eigenvalue problem
\begin{equation}
det|M^{-1} F - \lambda I| = 0
\end{equation}

But in the case of analytical solutions done with code this creates a few errors:
1. Simply put in a realistic system with variable masses, M, and variable bond strengths, C, within the force matrix there will not be nearly as many canceled terms in solving the determinant as when all masses or bonds are standardized. This leads to a shape mismatch with nearly all linear algebra solvers when trying to find eigenvalues.
2. Due to the physical nature of these matrices as not just being numerical, but rather physical operators they follow Hermitian properties. This means that while $ F = F^\dagger $ and $M = M^\dagger $, the value $M^{-1} F \neq (M^{-1} F)^\dagger $, so you cannot simply calculate this by moving the mass matrix over. Trying to do so leads to some solutions, but loses many of the smaller solutions critical to fully describing the system.
3. Most linear algebra solvers are programmed to solve the most general form $A \nu = \lambda \nu $ as mentioned above. Any variations of this simply break the system.  


#### So then how is it properly solved?  
The linear algebra solver in Scipy simplifies a more general matrix problem into the very basic form by taking the system 
\begin{equation}
F \nu = \lambda M \nu
\end{equation}
and applying some decomposition methods to turn it into a form
\begin{equation}
F^* \nu = \lambda \nu
\end{equation}
where $F^* = PLU$  

#### Firstly, What is LU Decomposition?  
For any matrix $A = M \times N$ where $A$ may or may not be square the matrix $A$ can be represented as $A=PLU$. This is a series of matrix multiplication where $P = M \times M$ matrix while $L$ and $U$ are the lower and upper triangular matrices of the forms $M \times K$ where $K = min(M,N)$ or $K=max(M,N)$.  
This is used to solve systems of matrices where on side has many solutions such as 
\begin{equation}
A x_i = b_i
\end{equation}
\begin{equation}
PLU x_i = b_i
\end{equation}

#### Secondly, What is Cholesky Decomposition?
Cholesky decomposition is a special case of LU decomposition applicable to Hermitian positive operators, $A=A^H$ and $x^H A x \geq 0$.  
In this variation $A = U^H U = LL^H $, it should be noted $L=U^H$ and vice versa.  

#### Lastly, How does this apply?
In the case of the phonon states in the form
\begin{equation}
F \nu = \lambda M \nu
\end{equation}
The matrix $M$ can be represented in the form $LL^\dagger$ where $L = \sqrt{m_i}$  
The variable matrix $\nu$ can also be defined using the transformation $y=L^\dagger \nu$ to $\nu = L^{-\dagger} y$ which gives the equation
\begin{equation}
F L^{-\dagger} y = \lambda LL^\dagger L^{-\dagger} y
\end{equation}
simplifying this becomes
\begin{equation}
F L^{-\dagger} y = \lambda L y
\end{equation}
\begin{equation}
L^{-1} F L^{-\dagger} y = \lambda y
\end{equation}
which is now in the same form as an LU Decompositon and can be solved as $A = L^{-1} F L^{-\dagger}$.  
While it may seem complex this is the path Scipy uses to determine the eigenvalues (ie the allowed frequencies) for the phonon states within the Brillouin Zone.  
via: https://docs.scipy.org/doc/scipy/tutorial/linalg.html#eigenvalues-and-eigenvectors 

### Symbolic creation of phonon matrix from fundamentals

```python
import numpy as np
import sympy as sp
import matplotlib.pyplot as plt

def N_atomic_symbolic():
    '''
    Creates an array for 1D chain of N variety atoms and their forces. 
    Array determinant then found to calculate the phonon dispersion.
    '''     
    # Asks user how many different atoms are in the chain and varifies it is a valid integer
    while True:
        N_input = input("How many atoms in the chain? ")
        # Check empty input
        if N_input.strip() == "":
            print("Error: Please enter a value.")
            continue
        # Check integer validity
        try:
            N = int(N_input)
            if N <= 0:
                print("Error: Number of atoms must be a positive integer.")
                continue
            break
        except ValueError:
            print("Error: Please enter a valid integer.")
            continue

    
    # For symbolic system need to create a mass for each N atoms
    masses = [] # Creating a list for the massses
    M = sp.symbols('M') # Defining M as a symbol for mass in sympy
    for atom in range(N):
        mass = f"{M}{atom+1}" #Uses an f string to comnbine str and int values into a str for the library keys
        masses.append(mass)
    print(f'The masses are {masses}')

    
    # Each mass needs a list to describe the forces it feels from nearest neighbors
    # Can create a dictionary for each mass and it's net forces
    net_forces = {}
    for mass in masses:
        # Defines the empty lists for forces
        forces = [0] * N # This creates a list of 0's the size of N. ie if N = 3, forces = [0, 0, 0]
        net_forces[mass] = forces # Assigns the key of each mass for the forces
    print(f'The net forces are {net_forces}')

    
    # Need to define a way to determine which spring coefficients for atomic bonds match the correct atoms
    # Can use the list of masses to represent the 1D chain
    #print(masses)
    # Sympy enumerating from 1 to N creating lists of the coefficients and equations, not a list
    bonds = list(sp.symbols(f"C1:{N+1}"))
    atoms = list(sp.symbols(f"U1:{N+1}"))
    print(f'The coefficients are {bonds}')
    print(f'The Wave Equations are {atoms}')

    k, a = sp.symbols('k a') # Sympy symbols for exponential terms
    omega = sp.symbols('omega')
    
    # Create a dictionary for symbolic expressions
    force_exprs = {}  # main dictionary keyed by mass
    for i, mass in enumerate(masses):
        # previous and next indices with periodic bounds
        prev_i = (i - 1) % N
        next_i = (i + 1) % N
        # bond symbols
        C_prev = bonds[prev_i]
        C_curr = bonds[i]
        # displacement symbols
        U_prev = atoms[prev_i]
        U_curr = atoms[i]
        U_next = atoms[next_i]
        # apply special phase factors for first and last atoms due to change in supercell
        if i == 0:
            # first atom: previous displacement gets exp(-i k a)
            U_prev = U_prev * sp.exp(-sp.I * k * a)
        if i == N - 1:
            # last atom: next displacement gets exp(i k a)
            U_next = U_next * sp.exp(sp.I * k * a)
        # symbolic force expression for this mass
        F_i = -C_prev*(-U_prev) - C_prev*U_curr - C_curr*U_curr - C_curr*(-U_next)
        # add the M_i * omega^2 * U_i term for the determinant matrix
        M_i = sp.symbols(mass)  # create a symbol for the mass, e.g., M1, M2...
        F_i += M_i * omega**2 * U_curr
        # store in a sub-dictionary keyed by mass
        force_exprs[mass] = {"force_expr": F_i}
        '''
    for mass, subdict in force_exprs.items():
        print(f"{mass}:")
        sp.pprint(subdict['force_expr'])
        print()  # extra line for spacing
        '''
    # Taking the expressions out of the dictionary and using Sympy collect to sort based on the function for each mass
    for mass, subdict in force_exprs.items():
        expr = subdict['force_expr']
        # collect terms based on U1, U2, ..., UN
        collected_expr = sp.collect(expr, atoms)
        # replace original expression with the collected one
        force_exprs[mass]['force_expr_collected'] = collected_expr       
    for mass, subdict in force_exprs.items():
        print(f"{mass}:")
        sp.pprint(subdict['force_expr_collected'])
        print()


    for mass, subdict in force_exprs.items():
        expr = subdict['force_expr_collected']  
        # Create a list of the coefficients for each U_i value
        coeffs = []
        for U in atoms:
            coeff = expr.coeff(U)  # extract coefficient of U for each expression
            coeffs.append(coeff)
        # Replace the zero list with these coefficients
        net_forces[mass] = coeffs     
    print("The net forces are:")
    for i, mass in enumerate(masses):
        print(f"({mass}): {net_forces[mass]}")
    

    # With each matrix in net_forces having the correct symbols for each term need to now combine it into one full matrix
    # Construct matrix rows in the order of masses
    matrix_rows = [net_forces[mass] for mass in masses]
    # Convert to SymPy Matrix
    D_matrix = sp.Matrix(matrix_rows)
    sp.pprint(D_matrix)


    # Solving for determinant here makes a mess of solutions so we limit the available wavevectors to helpful ranges
    # Long-wavelength limit
    k_small = 0  # Ka << 1, basically k = 0
    # Zone-boundary limits
    k_pi = sp.pi / a
    k_minus_pi = -sp.pi / a
    # Compute determinant
    det_expr = D_matrix.det() 
    #print("\nDeterminant of the force matrix:")
    #sp.pprint(det_expr)
    # Subbing in the limiting conditions
    det_k0 = det_expr.subs(k, k_small)
    det_kpi = det_expr.subs(k, k_pi)
    det_kminpi = det_expr.subs(k, k_minus_pi)
    # Solving the solutions of the determinant at these limits det = 0 for omega^2 
    solutions_k0 = sp.solve(det_k0, omega**2)
    solutions_kpi = sp.solve(det_kpi, omega**2)
    solutions_kminpi = sp.solve(det_kminpi, omega**2)
    # Print all symbolic solutions
    print("\nSymbolic solutions for ω^2 when K~0:")
    for sol in solutions_k0:
        sp.pprint(sol)
    # These work but are extremely messy
    
    print("\nSymbolic solutions for ω^2 when K=pi/a:")
    for sol in solutions_kpi:
        sp.pprint(sol)
    print("\nSymbolic solutions for ω^2 when K=-pi/a:")
    for sol in solutions_kminpi:
        sp.pprint(sol)
   

N_atomic_symbolic()
```

### Numeric Derivation of Phonon Branches

```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.linalg


def N_atomic_numeric():
    """
    Fully numerical version of the symbolic function.
    Builds the dynamical matrix for each k and returns ω(k) branches.
    """
    
    # Asks user how many different atoms are in the chain and varifies it is a valid integer
    while True:
        N_input = input("How many atoms in the chain? ")
        # Check empty input
        if N_input.strip() == "":
            print("Error: Please enter a value.")
            continue
        # Check integer validity
        try:
            N = int(N_input)
            if N <= 0:
                print("Error: Number of atoms must be a positive integer.")
                continue
            break
        except ValueError:
            print("Error: Please enter a valid integer.")
            continue
    
    
    masses = []  # Creates a list for all the masses to be used
    
    for atom in range(N):  # Takes user input of mass for each atom and adds to list of masses while checking it is a valid mass
        while True:
            mass_input = input(f"What is the mass of atom {atom + 1}? ")
            # Empty check
            if mass_input.strip() == "":
                print("Error: Please enter a value.")
                continue
            # Numeric check
            try:
                mass = float(mass_input)
                break
            except ValueError:
                print("Error: Mass must be a valid number (integer or decimal).")
                continue
        masses.append(mass)
    print(f"The atomic masses are {masses}")

    
    # Need to define a way to determine which spring coefficients for atomic bonds match the correct atoms
    # Can create a list of all the bonds and then ask the user for an exact value to replace the variable
    bonds = []
    for atom in range(N):
        if atom < (N-1):
            bonds.append(f"C{atom+1}{atom+2}")
        else:
            bonds.append(f"C{atom+1}{N-(N-1)}")
    print(f'The coefficients are {bonds}') # Prints a nice list of the coefficients representing their connections
 

    # To determine the bond calculations the first step would be to ask if the user wants a simplified or realistic case
    # Simplified being C12 = C23 = C34... CMN
    # Can concurrently do the same for unit cell spacing, a
    while True:
        bond_input = input("Would you like simplified or realistic calculations? Note: Simplified means all like terms equal ")
        # Empty check
        if bond_input.strip() == "":
            print("Error: Please enter an answer.")
            continue
            
        # Accepting a variety of answers based on possible expectations
        simplified = ['1', 'simplified', 'Simplified', 's', 'simple', 'Simple']
        realistic = ['2', '0', 'realistic', 'Realistic', 'real', 'Real']
        
        # Just need to ask for unit cell spacing once as it is unique per crystal not atom
        try:
            a = float(input('What is the atomic spacing? '))
        except ValueError:
            print("Error: atomic spacing must be a number.")
            continue
        print(f' The unit cell spacing is {a}')
        
        # Now to actually determine the bond values
        if bond_input in simplified:
        # If simplified then all bonds will be the same
            try:
                bond_val = float(input('What is the uniform bond value? '))
            except ValueError:
                print("Error: bond value must be numeric.")
                continue
            for i,bond in enumerate(bonds):
                bonds[i] = bond_val
            print(f'The atomic bond strengths are {bonds}')
            break
        # If realistic each bond needs a numeric value    
        elif bond_input in realistic:
            for i,bond in enumerate(bonds):
                try:
                    bond_val = float(input(f'What is the bond strength of {bond}? '))
                except ValueError:
                    print("Error: bond value must be numeric.")
                    break
                bonds[i] = bond_val
            print(f'The atomic bond strengths are {bonds}')
            break
        else:
            print('Invalid Option')


    # k values over the Brillouin zone
    kpoints = 400
    k_vals = np.linspace(-np.pi/a, np.pi/a, kpoints)

    omega_vals = np.zeros((kpoints, N))

    for ki, k in enumerate(k_vals):
        # Special cases for N = 1 and N = 2 are needed as the indexing used for N > 2 has overlapping commands for N < 2
        # Using the standard versions guaranteees accuracy
        if N == 1:
            """
            Monatomic chain dispersion:
                omega(k) = 2 * sqrt(C/m) * |sin(k*a/2)|
            """
            C = bonds[0]
            m = masses[0]
            omega_vals[ki, 0] = 2.0 * np.sqrt(C / m) * np.abs(np.sin(k * a / 2.0))
    
        elif N == 2:
            '''
            Diatomic Chain With Variable Masses and Bonds:

                omega^2 = (C1+C2)*(m1+m2)/(2*m1*m2) ± sqrt([(C1+C2)(m1+m2)]^2 - 16(m1*m2)(C1*C2)*sin^2(ka/2))/(2*m1*m2)
            '''
            m1, m2 = masses
            C1, C2 = bonds
           
            s = np.sin(k * a / 2.0)

            prefactor = (C1 + C2) * (m1 + m2) / (2.0 * m1 * m2)
            
            disc = ((C1 + C2) * (m1 + m2))**2 \
                   - 16.0 * m1 * m2 * C1 * C2 * s**2
            
            disc = np.maximum(disc, 0.0)
            root = np.sqrt(disc) / (2.0 * m1 * m2)
            
            omega_sq_plus  = prefactor + root
            omega_sq_minus = prefactor - root
            
            omega_plus  = np.sqrt(np.maximum(omega_sq_plus, 0.0))
            omega_minus = np.sqrt(np.maximum(omega_sq_minus, 0.0))
            
            omega_vals[ki, :] = np.sort([omega_minus, omega_plus])

        else:
            M = np.diag(masses)  # mass matrix for generalized eigenproblem
            
            # Build physical dynamical matrix (no mass normalization)
            D = np.zeros((N, N), dtype=complex)
            for i in range(N):
                prev_i = (i - 1) % N
                next_i = (i + 1) % N
    
                C_prev = bonds[prev_i]
                C_curr = bonds[i]
    
                # Diagonal term: sum of spring constants connected to this atom
                D[i, i] = C_prev + C_curr
    
                # Off-diagonal: previous neighbor
                D[i, prev_i] = -C_prev
                if i == 0:  # Bloch phase for periodic boundary
                    D[i, prev_i] *= np.exp(-1j * k * a)
    
                # Off-diagonal: next neighbor
                D[i, next_i] = -C_curr
                if i == N - 1:  # Bloch phase for periodic boundary
                    D[i, next_i] *= np.exp(1j * k * a)
    
            # Solve generalized eigenproblem D U = ω^2 M U
            w2, _ = scipy.linalg.eigh(D, M)
            w2 = np.real_if_close(w2, tol=1000)
            w2[w2 < 0] = 0.0
    
            # Sort and take sqrt to get ω
            omega_vals[ki, :] = np.sqrt(np.sort(w2))

    
    return omega_vals, k_vals, a, masses, bonds
    
```

```python
def plot_phonon_branches(k_vals, omega_vals, a):
    '''
    Plots phonon dispersion branches for a 1D N-atom chain
    '''
    N = omega_vals.shape[1]  # number of branches

    fig, ax = plt.subplots(figsize=(8, 5))

    # Plot each branch
    for branch in range(N):
        if branch == 0:
            label = "Acoustic"
        else:
            label = f"Optical {branch}"  # Optical branches numbered 1,2,3...
        ax.plot(k_vals, omega_vals[:, branch], lw=2, label=label)


    # Title and labels
    ax.set_title(f"1D chain with {N} atoms per unit cell")
    ax.set_xlabel("wave vector, k")
    ax.set_ylabel("frequency, ω")

    ax.set_xlim(-np.pi/a, np.pi/a)
    ax.set_xticks([-np.pi/a, -np.pi/(2*a), 0, np.pi/(2*a), np.pi/a])
    ax.set_xticklabels([r"$-\pi/a$", r"$-\pi/2a$", "0", r"$\pi/2a$", r"$\pi/a$"])

    ax.grid(True)
    # Legend in the margin to the right
    ax.legend(loc='center left', bbox_to_anchor=(1, 0.5))
    plt.tight_layout()
    plt.show()
```

```python
omega_vals, k_vals, a, _, _ = N_atomic_numeric()
plot_phonon_branches(k_vals, omega_vals, a)
```

```python
import numpy as np
import matplotlib.pyplot as plt

def draw_atomic_chain(a, masses, bonds, n_cells=3):
    '''
    Draws a 1D atomic chain with periodicity.
    '''
    masses = np.array(masses)
    bonds = np.array(bonds)
    N = len(masses)  # atoms per unit cell

    # spacing between atoms inside unit cell
    dx = a / N  

    # Repeat unit cell
    masses_rep = np.tile(masses, n_cells)
    bonds_rep = np.tile(bonds, n_cells)
    Ntot = len(masses_rep)

    # x positions
    x = np.arange(Ntot) * dx
    y = np.zeros(Ntot)

    # Normalize atom sizes
    m_min, m_max = masses.min(), masses.max()
    sizes = 300 + 1200 * (masses_rep - m_min) / (m_max - m_min + 1e-12)

    # Normalize bond widths
    C_min, C_max = bonds.min(), bonds.max()
    widths = 1.5 + 6.0 * (bonds_rep - C_min) / (C_max - C_min + 1e-12)

    fig, ax = plt.subplots()

    # Draw bonds
    for i in range(Ntot - 1):
        ax.plot(
            [x[i], x[i + 1]],
            [y[i], y[i + 1]],
            lw=widths[i % N],
            color="black",
            solid_capstyle="round")

    # Draw atoms
    ax.scatter(x, y, s=sizes, color="steelblue", zorder=3)

    # Draw scale bar for the first unit cell only
    bar_y = min(y) - 0.2 * a
    ax.plot([0, a], [bar_y, bar_y], color='red', lw=3)
    ax.text(a/2, bar_y - 0.05*a, "a", ha='center', va='top', color='red', fontsize=12)

    # Set limits to prevent clipping
    ax.set_xlim(x[0] - dx, x[-1] + dx)
    ax.set_ylim(bar_y - 0.1*a, max(y) + 0.2*a)

    ax.set_aspect("auto")
    ax.axis("off")
    ax.set_title(f"1D Atomic Chain ({n_cells} Unit Cells)")

    plt.show()
```

```python
_, _, a, masses, bonds = N_atomic_numeric()
draw_atomic_chain(a, masses, bonds, n_cells=3)
```
